# MCDemo

Monte Carlo Demo


```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
n = 1_000_000 # # Scenarios
S = 8
K = 10
sigma = .2
T = 1.5
r = 0.05
```

```python
S_T = S * np.exp(((r-sigma**2/2.0)*T) + (sigma * np.sqrt(T) * np.random.normal(0, 1, size=n)))
C = np.maximum(S_T - K, 0).mean()*np.exp(-r*T)
C
```

```python
d1 = (np.log(S/K) + (r+sigma**2/2.0)*T)/(sigma*np.sqrt(T))
d2 = d1 - sigma*np.sqrt(T)
C_bs = S*st.norm.cdf(d1) - K*np.exp(-r*T)*st.norm.cdf(d2)
C_bs
```
